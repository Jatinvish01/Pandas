# Querying with .query()

#### The .query() method in pandas lests you filter DataFrame row using a string expression -- it's a more readable and often more concise alternative to using boolean indexing

```python
# This is a cleaner, SQL-like way to filter:
df.query("Year > 2018 and IMDb > 6")
```

#### You can use @ TO reference Python varibables

```python
year_limit = 2018
df.query("Year > @year_limit")
```
